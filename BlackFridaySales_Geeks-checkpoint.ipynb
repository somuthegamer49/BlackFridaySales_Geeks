```python
df['Gender'].value_counts().plot(kind='bar',title="Gender Count",figsize=(10,5))
```

## More men are doing shopping as compared to women

```python
plt.figure(figsize=(10,5))
df.groupby('Gender')['Purchase'].sum().plot(kind='bar')
plt.ylabel('Total Purchase')
```

## Men have contributed to high revenue as compared to women

```python
plt.figure(figsize=(10,5))
gen_marital=df.groupby('Gender')['Marital_Status'].value_counts()
gen_marital=pd.DataFrame(gen_marital.reset_index(name='count'))
sns.barplot(x=gen_marital['Gender'],y=gen_marital['count'],hue=gen_marital['Marital_Status'])
```

## Unmarried people are purchasing more as compared to married ones. Also, Men are purchasing more in adding to more sales

```python
plt.figure(figsize=(10,5))
agewise_gender=pd.DataFrame(df.groupby('Age')['Gender'].value_counts().reset_index(name='Count'))
sns.barplot(x=agewise_gender['Age'],y=agewise_gender['Count'],hue=agewise_gender['Gender'])
```

## Age 18 to 45 have contributed to more sales. Age group 26 to 35 are contributing highest. 

```python
df.groupby('City_Category')['Purchase'].sum().plot(kind='pie',autopct="%.2f",title="Citywise Purchase Amount",figsize=(10,5))
```

## City B has contributed to highest revenue followed by C and then A

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(10, 5),squeeze = False) 
x=df.groupby('Age')['Purchase'].sum()
y=df.groupby('Marital_Status')['Purchase'].sum()

axes[0, 0].pie(x,labels=df['Age'].unique(),autopct='%.2f')
axes[0, 1].pie(y,labels=df['Marital_Status'].unique(),autopct='%.2f')
axes[0,0].set_title('Age Wise Total Purchase Distribution')
axes[0,1].set_title('Marriage Wise Total Purchase Distribution')
plt.show() 
```

## Age 26-35 has contributed to highest revenue followed by 46-50 and then 55+. Unmarried people have contributed to most revenue

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['City_Category'],hue=df['Age'])
plt.show()
```

## Age group 18-25 from city A and C have same shopping frequency. City B has more frequency in terms of purchase

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['Gender'],hue=df['Age'])
plt.show()
```

## Men from 26 to 35 age group have contributed to highest sales followed by age 36 to 45 and then 18 to 25

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['Product_Category_1'],hue=df['City_Category'])
plt.show()
```

## Product category 1, 5 and 8 has highest sales in the order of city B,C and A

```python
plt.figure(figsize=(15,6))
data1=pd.DataFrame(df.groupby('Occupation')['City_Category'].value_counts().reset_index(name="Count"))
data1=data1.pivot('City_Category','Occupation','Count')
sns.heatmap(data=data1)
plt.show()
```

## People with occupation 4,0 and 7 have contributed highly to sales. City B has more sales contribution

```python
plt.figure(figsize=(15,6))
data1=pd.DataFrame(df.groupby('Occupation')['Product_ID'].value_counts().reset_index(name="Count"))
data1=data1.pivot('Product_ID','Occupation','Count')
sns.heatmap(data=data1)
plt.show()
```

## People with Occupation 8 has purchased all sorts of product that too with high frequency

```python
cols=['Gender','Age','Occupation','City_Category']
plt.figure(figsize=(15,6))
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(10, 5),squeeze=False) 
sns.countplot(x=df['Stay_In_Current_City_Years'],hue=df[cols[0]],ax=axes[0,0])
sns.countplot(x=df['Stay_In_Current_City_Years'],hue=df[cols[3]],ax=axes[0,1])
plt.show() 
```

## Most people stay in a city for 1 year only that too men. Most of the buyers are men who are living in a city for short term.

```python
plt.figure(figsize=(15,6))
df_pr=df['Product_ID'].value_counts().sort_values(ascending=False).reset_index()
df_pr.columns=['Product_ID','Count']
df_pr=df_pr.head(10)
sns.barplot(x=df_pr['Product_ID'],y=df_pr['Count'])
plt.show()
```
